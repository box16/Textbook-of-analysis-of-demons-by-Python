```python
import matplotlib.pyplot as pyplot
import matplotlib.style
matplotlib.style.use("ggplot")
```

### 折れ線グラフ
plotメソッドは折れ線グラフを描画する

```python
x = [i for i in range(10)]
y_1 = [i**2 for i in x]
y_2 = x

canvas,sub_plot = pyplot.subplots()
sub_plot.plot(x,y_1,label="y=x**2")
sub_plot.plot(x,y_2,label="y=x")


sub_plot.legend(loc="best")

sub_plot.set_xlabel("x")
sub_plot.set_ylabel("y")

pyplot.show()
```

折れ線グラフでも値の間隔が狭ければ曲線に近似できる

```python
import numpy
x = numpy.arange(0,15,0.1)
y_1 = numpy.sin(x)
y_2 = numpy.cos(x)

canvas,sub_plot = pyplot.subplots()
sub_plot.plot(x,y_1,label="sin")
sub_plot.plot(x,y_2,label="cos")


sub_plot.legend(loc="best")

sub_plot.set_xlabel("x")
sub_plot.set_ylabel("y")

pyplot.show()
```

### 棒グラフ
棒グラフの描画にはbarメソッドを用いる

```python
x = [i for i in range(10)]
y_1 = [i**2 for i in x]
y_2 = x

canvas,sub_plot = pyplot.subplots()
sub_plot.bar(x,y_1,label="y=x**2")
sub_plot.bar(x,y_2,label="y=x")


sub_plot.legend(loc="best")

sub_plot.set_xlabel("x")
sub_plot.set_ylabel("y")

pyplot.show()
```

```python
x = [i for i in range(10)]
y_1 = [i**2 for i in x]

canvas,sub_plot = pyplot.subplots()

# tick_labelを指定すれば、それぞれの棒にラベルを付けることも出来る
labels = [f"{i}" for i in y_1]
sub_plot.bar(x,y_1,tick_label=labels)


sub_plot.set_xlabel("x")
sub_plot.set_ylabel("y")

pyplot.show()
```

```python
x = [i for i in range(10)]
y_1 = [i**2 for i in x]
y_2 = x

# barhで横向きの棒グラフを描画できる
canvas,sub_plot = pyplot.subplots()
sub_plot.barh(x,y_1,label="y=x**2")
sub_plot.barh(x,y_2,label="y=x")


sub_plot.legend(loc="best")

sub_plot.set_xlabel("x")
sub_plot.set_ylabel("y")

pyplot.show()
```

二つの棒グラフを並べて表示する場合は少し面倒

```python
x = [i for i in range(10)]
y_1 = [i**2 for i in x]
y_2 = x

# まず、棒の幅を決める
width = 0.4

canvas,sub_plot = pyplot.subplots()
sub_plot.bar(x,y_1,label="y=x**2",width=width)

# ずらす側を調整する
x_2 = [i+0.4 for i in x]
sub_plot.bar(x_2,y_2,label="y=x",width=width)


sub_plot.legend(loc="best")

sub_plot.set_xlabel("x")
sub_plot.set_ylabel("y")

pyplot.show()
```

積み上げタイプの棒グラムも上記同様、ちょっと手間を加える必要がある

```python
x = [i for i in range(10)]
y_1 = [i**2 for i in x]

canvas,sub_plot = pyplot.subplots()

# 先に大きいほうを描画する
y_2 = [i+value for i,value in enumerate(y_1)]
sub_plot.bar(x,y_2,label="y=x")

sub_plot.bar(x,y_1,label="y=x**2")

sub_plot.legend(loc="best")

sub_plot.set_xlabel("x")
sub_plot.set_ylabel("y")

pyplot.show()
```

### 円グラフ
pieメソッドで円グラフを作成できる

```python
labels = ["A","B","C"]
x = numpy.random.randint(0,10,3)

canvas,sub_plot = pyplot.subplots()
sub_plot.pie(x,labels=labels)

pyplot.show()
```

```python
labels = ["A","B","C"]
x = numpy.random.randint(0,10,3)

canvas,sub_plot = pyplot.subplots()
sub_plot.pie(x,labels=labels)

# アスペクト比を固定して円形を保つ
sub_plot.axis("equal")

pyplot.show()
```

円グラフはデフォルトでは3時の方向から、反時計回りに描画される  
12時の方向から時計回りに描画するには調整しないといけない

```python
labels = ["A","B","C"]
x = numpy.random.randint(0,10,3)

canvas,sub_plot = pyplot.subplots()
# X,Y座標をイメージしているため、3時方向が0,12時方向が90度となる
sub_plot.pie(x,labels=labels,startangle=90,counterclock=False)

sub_plot.axis("equal")

pyplot.show()
```

```python
labels = ["A","B","C"]
x = numpy.random.randint(0,10,3)

canvas,sub_plot = pyplot.subplots()
# shadowで影付き
# autopctでパーセント表示が可能
sub_plot.pie(x,labels=labels,startangle=90,counterclock=False,
             shadow=True,autopct="%1.2f%%")

sub_plot.axis("equal")

pyplot.show()
```

```python
labels = ["A","B","C"]
x = numpy.random.randint(0,10,3)

canvas,sub_plot = pyplot.subplots()

# explodeで、切り出したい要素に何かしらの値を入れると、切り出してくれる
# 今回の場合はBを0.2分切り出す
explode = [0,0.2,0]
sub_plot.pie(x,labels=labels,startangle=90,counterclock=False,
             shadow=True,autopct="%1.2f%%",
             explode=explode)

sub_plot.axis("equal")

pyplot.show()
```

### 複数のグラフを組み合わせる

```python
x = [i for i in range(10)]
y_1 = numpy.random.randint(0,30,len(x))
y_2 = numpy.random.randint(0,30,len(x))
y_3 = numpy.random.randint(0,30,len(x))

canvas,sub_plot = pyplot.subplots()
sub_plot.plot(x,y_1,label="random1")
sub_plot.scatter(x,y_2,label="random2")
sub_plot.bar(x,y_3,label="random3")

sub_plot.legend(loc="best")

sub_plot.set_xlabel("x")
sub_plot.set_ylabel("y")

pyplot.show()
```

これを応用すれば、ヒストグラムに正規分布の近似曲線をプロットすることも可能

```python
mu = 100
sigma = 15

x = numpy.random.normal(mu,sigma,1000)
canvas,sub_plot = pyplot.subplots()

n,bins,patches = sub_plot.hist(x,bins=30)

# 正規分布に必要なx側の点を求める
# この場合、ヒストグラムの中点に位置する値で良い
# bins[:-1]+bins[1:]とすることで、各階級の下限と上限の加算のリストになる
# それを平均することで、各階級値が求められる
fit_x = (bins[:-1]+bins[1:])/2

# 各xに対して正規分布上のyの値を求める
# 正規分布の式
fit_y = numpy.exp(-(fit_x-mu)**2/(2*sigma**2))/(numpy.sqrt(2*numpy.pi)*sigma)
# この状態のfit_yはただの出現確率
# これを度数にまで持っていく
# 試行回数は1000回であるため、*1000でそれぞれの出現数となる
# ヒストグラムには幅があるため、その分加算できる
fit_y = fit_y*1000*numpy.diff(bins)
sub_plot.plot(fit_x,fit_y)

pyplot.show()
```
