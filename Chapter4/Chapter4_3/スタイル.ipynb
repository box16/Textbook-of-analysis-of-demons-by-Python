```python
import matplotlib.pyplot as pyplot
```

### 色の設定

```python
x = [i for i in range(10)]

canvas,sub_plot = pyplot.subplots()

# 色名で指定
y_1 = [i**2 for i in x]
sub_plot.plot(x,y_1,label="aqua",color="aqua")

# カラーコードで指定
# もう二桁を最後に追加して透明度を指定できる
y_2 = [i**3 for i in x]
sub_plot.plot(x,y_2,label="#000FFF",color="#000FFF")

# RGBAをfloatで指定
y_3 = [i**4 for i in x]
sub_plot.plot(x,y_3,label="(0.1,0.2,0.5,0.3)",color=(0.1,0.2,0.5,0.3))

sub_plot.legend(loc="best")

sub_plot.set_xlabel("x")
sub_plot.set_ylabel("y")

pyplot.show()
```

基本的に、HTMLやCSSで指定するような色名、カラーコードが使用できる  
また、棒グラフの場合は、枠線と、面積部分の色をそれぞれ指定できる

```python
x = [i for i in range(10)]

canvas,sub_plot = pyplot.subplots()

# 塗りつぶしと枠線をそれぞれ指定できる
y_1 = [i**2 for i in x]
sub_plot.bar(x,y_1,label="aqua",color="aqua",edgecolor="red")

sub_plot.legend(loc="best")

sub_plot.set_xlabel("x")
sub_plot.set_ylabel("y")

pyplot.show()
```

## 線のスタイル

```python
x = [i for i in range(10)]

canvas,sub_plot = pyplot.subplots()

# ライン幅を5.5で指定
y_1 = [i**2 for i in x]
sub_plot.plot(x,y_1,label="5.5",linewidth=5.5)

y_2 = [i**3 for i in x]
sub_plot.plot(x,y_2,label="10",linewidth=10)

sub_plot.legend(loc="best")

sub_plot.set_xlabel("x")
sub_plot.set_ylabel("y")

pyplot.show()
```

```python
x = [i for i in range(10)]

canvas,sub_plot = pyplot.subplots()

# 破線
y_1 = [i**2 for i in x]
sub_plot.plot(x,y_1,label="dashed",linestyle="--")

# 一点鎖線
y_2 = [i**3 for i in x]
sub_plot.plot(x,y_2,label="dashdot",linestyle="-.")

# 点線
y_3 = [i**4 for i in x]
sub_plot.plot(x,y_3,label="dotted",linestyle=":")

sub_plot.legend(loc="best")

sub_plot.set_xlabel("x")
sub_plot.set_ylabel("y")

pyplot.show()
```

### フォント

```python
x = [i for i in range(10)]

canvas,sub_plot = pyplot.subplots()

y_1 = [i**2 for i in x]
sub_plot.plot(x,y_1,label="5.5",linewidth=5.5)

sub_plot.legend(loc="best")

sub_plot.set_xlabel("x-label",family="fantasy",size=20,weight="bold")
sub_plot.set_ylabel("y-label",family="cursive",size=40,weight="light")
sub_plot.set_title("graph title",family="monospace",size=25,weight="heavy")

pyplot.show()
```

```python
x = [i for i in range(10)]

canvas,sub_plot = pyplot.subplots()

y_1 = [i**2 for i in x]
sub_plot.plot(x,y_1,label="5.5",linewidth=5.5)

sub_plot.legend(loc="best")

# 辞書型でフォントを指定できる
font_dic = {"family":"fantasy","size":20,"weight":"bold"}

sub_plot.set_xlabel("x-label",fontdict=font_dic)
sub_plot.set_ylabel("y-label",fontdict=font_dic)
sub_plot.set_title("graph title",fontdict=font_dic)

pyplot.show()
```

### テキスト描画
グラフに任意のテキストを表示させる機能  
X,Y座標を指定する必要がある

```python
x = [i for i in range(10)]

canvas,sub_plot = pyplot.subplots()

y_1 = [i**2 for i in x]
sub_plot.plot(x,y_1,label="5.5",linewidth=5.5)
sub_plot.text(1,60,"text",size=20)

sub_plot.legend(loc="best")


pyplot.show()
```

### pandasのオブジェクトからグラフ描画

```python
import pandas

data_frame = pandas.DataFrame({"A":[1,2,3],"B":[4,5,6]})
data_frame.plot()
pyplot.show()
```

```python
data_frame = pandas.DataFrame({"A":[1,2,3],"B":[4,5,6]})
# pandasのbarは複数ある場合自動で横並びにしてくれる
data_frame.plot.bar()
pyplot.show()
```

```python
data_frame = pandas.DataFrame({"A":[1,2,3],"B":[4,5,6]})
# pandasのbarは積み上げも簡単に行える
data_frame.plot.bar(stacked=True)
pyplot.show()
```
