### Matplotlibとは
Matplotlibとは、グラフ描画のためのライブラリ  
jupyter notebookとの親和性が高く、コードとデータとグラフが同じドキュメントで表示されるため便利  
  
Matplotlibでグラフを描画するには2つのスタイルがある  
  
* MATLABスタイル
* オブジェクト指向スタイル
  
本書ではオブジェクト指向スタイルで描画する

```python
import matplotlib.pyplot as pyplot
```

```python
# グラフスタイルは"ggplot"とする※後述
import matplotlib.style
matplotlib.style.use("ggplot")
```

### MATLABスタイルでの描画
MATLABという数値解析ソフトウェアと似た形式でグラフ描画を行う方法

```python
x = [1,2,3]
y = [2,4,9]

# グラフに用いるデータの指定(2次元)
pyplot.plot(x,y)
# グラフタイトルの指定
pyplot.title("MATLAB-style")
# グラフ描画
pyplot.show()
```

### オブジェクト指向スタイルでの描画
オブジェクト指向スタイルでは、  
まず、描画オブジェクトに対して、サブプロットを追加する  
そして、サブプロットに対して描画を行う  
  
これによって、複数のグラフをまとめて表示するといったことが可能になる  
つまり拡張性がオブジェクト指向スタイルの利点

```python
x = [1,2,3]
y = [2,4,9]

# 描画オブジェクト(大枠)とサブプロット(子要素)を生成
canvas,sub_plot = pyplot.subplots()

sub_plot.plot(x,y)
sub_plot.set_title("Object-style")

pyplot.show()
```
