### 描画オブジェクトとサブプロット
Matplotlibでは、**1つの描画オブジェクトに,1つ以上のサブプロット**を配置する  
subplots関数に引数を渡すとその分のサブプロットが返ってくる

```python
import matplotlib.pyplot as pyplot
import matplotlib.style
matplotlib.style.use("ggplot")
```

```python
x = [1,2,3]
y_1 = [i**4 for i in x]
y_2 = x

canvas,sub_plots = pyplot.subplots(2)
sub_plots[0].plot(x,y_1)
sub_plots[1].plot(x,y_2)
pyplot.show()
```

```python
# 行列のように複数作成することも可能
canvas,sub_plots = pyplot.subplots(2,2)
pyplot.show()
```

```python
# nrows, ncolsでどういった構成にするかは指定できる
canvas,sub_plots = pyplot.subplots(nrows=2,ncols=3)
pyplot.show()
```

### グラフのスタイル
グラフスタイルとは、**線の色、太さ、背景色**などがある程度決まったプリセットのような者  
`matplotlib.style.available`で使用可能なスタイルが取得できる

```python
print(matplotlib.style.available)
print(len(matplotlib.style.available))
```

```python
matplotlib.style.use("Solarize_Light2")
x = [i for i in range(10)]
y = [i**4 for i in x]

canvas,sub_plot = pyplot.subplots()
sub_plot.plot(x,y)
sub_plot.set_title("Solarize_Light2")
pyplot.show()
```

```python
matplotlib.style.use("seaborn-pastel")
x = [i for i in range(10)]
y = [i**4 for i in x]

canvas,sub_plot = pyplot.subplots()
sub_plot.plot(x,y)
sub_plot.set_title("seaborn-pastel")
pyplot.show()
```

```python
matplotlib.style.use("classic")
x = [i for i in range(10)]
y = [i**4 for i in x]

canvas,sub_plot = pyplot.subplots()
sub_plot.plot(x,y)
sub_plot.set_title("classic")
pyplot.show()
```

### タイトル
描画オブジェクトとサブプロットにはそれぞれタイトルを指定できる  

```python
matplotlib.style.use("ggplot")

x = [i for i in range(10)]
y_1 = [i**2 for i in x]
y_2 = x

canvas,sub_plots = pyplot.subplots(2)
sub_plots[0].plot(x,y_1)
sub_plots[1].plot(x,y_2)

sub_plots[0].set_title("y=x**2")
sub_plots[1].set_title("y=x")

canvas.suptitle("Canvas")
pyplot.show()
```

### 軸ラベル

```python
x = [i for i in range(10)]
y = [i**2 for i in x]

canvas,sub_plot = pyplot.subplots()
sub_plot.plot(x,y)

sub_plot.set_title("y=x**2")

sub_plot.set_xlabel("x")
sub_plot.set_ylabel("y")

pyplot.show()
```

### 凡例
凡例の表示はplot時に、label引数に凡例名を指定する  
legendメソッドで凡例の描画方法を指定できる  
loc="best"はデータとの重なりが最小になるように出力してくれる

```python
x = [i for i in range(10)]
y_1 = [i**2 for i in x]
y_2 = x

canvas,sub_plot = pyplot.subplots()
sub_plot.plot(x,y_1,label="y=x**2")
sub_plot.plot(x,y_2,label="y=x")


sub_plot.legend(loc="best")

sub_plot.set_xlabel("x")
sub_plot.set_ylabel("y")

pyplot.show()
```

```python
x = [i for i in range(10)]
y_1 = [i**2 for i in x]
y_2 = x

canvas,sub_plot = pyplot.subplots()
sub_plot.plot(x,y_1,label="y=x**2")
sub_plot.plot(x,y_2,label="y=x")

# 右下に凡例を表示させる
sub_plot.legend(loc="lower right")

sub_plot.set_xlabel("x")
sub_plot.set_ylabel("y")

pyplot.show()
```

### ファイル出力
作成したグラフは画像形式で保存できる  
ファイル名で直接拡張子を指定する

```python
x = [i for i in range(10)]
y_1 = [i**2 for i in x]
y_2 = x

canvas,sub_plot = pyplot.subplots()
sub_plot.plot(x,y_1,label="y=x**2")
sub_plot.plot(x,y_2,label="y=x")


sub_plot.legend(loc="best")

sub_plot.set_xlabel("x")
sub_plot.set_ylabel("y")

canvas.savefig("function_sample.png")
```
