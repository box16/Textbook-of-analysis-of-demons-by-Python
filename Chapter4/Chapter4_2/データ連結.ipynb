```python
import numpy
import pandas
```

### DataFrame同士の連結

```python
health_data = pandas.read_pickle("health_data_20210327.pickle")
health_data
```

```python
exercise_data = pandas.read_pickle("one-hot_exercise.pickle")
exercise_data
```

### 列方向に連結
複数のDataFrameを結合し、横方向(列方向)に大きくする  
結合の際は、indexが同じもの同士で結合される

```python
# 結合はconcat
# axis=1が列方向を指す
merge_column_data = pandas.concat([health_data,exercise_data],axis=1)
merge_column_data
```

### 行方向に連結

```python
# 結合はconcat
# axis=0が行方向を指す
# 存在しないデータはNaNで埋められる
merge_row_data = pandas.concat([health_data,exercise_data],axis=0,sort=True)
merge_row_data
```
