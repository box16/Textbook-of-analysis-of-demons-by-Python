### 解析学
関数の微分と積分が重要な要素となる分野  
物理法則と深く結びついている

### 積分は面積
次の式を考える  
$y = x$
そして、$(a,0)$となるような線を引き、囲まれる領域の面積を求めたいとする  
  
  
この時、面積の形は直角三角形であるため、底辺a  
高さ$y=x,(x=a) \dots y=a$で高さa  
となるため、面積を表現する関数Fは次のようになる  
$F(a) = \frac{1}{2}a^2$  
  
この流れを積分を用いて表現することが出来る  
$F(a) = \int_0^a xdx = \frac{1}{2}a^2$  
この場合、aは0以上任意の数までを取り得る  
こういった、特定の範囲で積分することが決まっているものを**定積分**という  
dxはxで積分することを意味する  

### 微分は傾き
$F(a) = \frac{1}{2}a^2$  
これを用いる  
a軸上にある点を取り、その座標をxとする  
そこにh分加算した位置に新しい点を取り、これら2点を通る直線を考える  
  
この直線の傾きを考えてみる  
傾きは  
$\frac{y方向の変化量}{x方向の変化量}$で求める事が出来る  
  
  
ここで、加算するhを極限まで小さくするとする。(0に近づく)  
すると、これで描かれる直線は、座標xにおけるF(a)への接線となる  
hを0とみなして計算すると次のようになる  
$\frac{1}{2}h+x \simeq x$  
つまり、点aを取った場所に応じて傾きは変わる  
$F(x) = x$  
これは、積分の時に扱った元の関数になっている  
  
面積を求める式F(x)は、傾きを求める関数f(x)に対して**原始関数**と呼ばれる  
一方、f(x)はF(x)の**導関数**と呼ばれる  
  
F(x)の微分は$F'(x)$と表現される  
もしくは
$\frac{dy}{dx}$  
と表現される
