2つ以上の変数を持つ、多変数関数に対しても微分を行うことはできる  
例えば以下の式  
$f(x,y) = \sin x \cos y$  
  
  
こういった複数の変数を含む関数を微分する場合は、変数毎に微分を分けて行う必要がある  
$\frac{\sigma f}{\sigma x} = \cos x \cos y$  
※$\sin x$のxについての微分は$\cos x$になる  
まず、上記のf(x,y)をxについて微分すると次のようになる  
この時、xを含まない$\cos y$の方は、定数であるとみなす  
  
また、yについての微分を次のように表すこともある  
$f_y = -\sin x \sin y$  
※$\cos y$のyについての微分は$-\sin y$となる
