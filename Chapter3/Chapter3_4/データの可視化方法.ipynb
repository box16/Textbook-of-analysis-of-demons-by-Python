### ヒストグラム
度数分布表を棒グラフで表現したものが**ヒストグラム**である  
  
ヒストグラムはデータの分布がどのような形になっているのかを把握できる優秀な図である  
  
これだけで、得られたデータが正規分布をしているのか否か程度の判断ができる

### 箱ひげ図
箱ひげ図は別名ボックスプロット(box plot)とも呼ばれる  
グループ間で分布に差があるのかを比較するときに便利な可視化方法  
  
  
中央の線が中央値で、広い意味では平均的な代表値を指す  
箱の下端、上端がそれぞれ、第一四分位数、第三四分位数で、箱の大きさがデータの50%分を占めている  
  
髭の部分は幾つか方法があり、  
1. 最大値、最小値までを髭とする(外れ値がない)
2. IQRを1.5倍した値を髭として、それ以外は外れ値とする
3. 第一四分位数～中央値までの大きさの1.5倍を下側の髭とし、中央値～第三四分位数までの大きさの1.5倍を上側の髭とする。それ以外は外れ値  


### 散布図
あるサンプルが二種類以上のデータによって示されるとき、散布図(scatter plot)を描くことが出来る  
  
上記の図の場合は、一つの点が1つの都道府県と対応する  
納豆の消費が多い都道府県では、わかめの消費も多い事が分かる  
(ただし、これは因果を示すものではない)
