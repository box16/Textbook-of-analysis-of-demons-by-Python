### 確率変数と確率分布
サイコロの例を考える  
全事象は  
  
$U = \{1,2,3,4,5,6\}$  
  
この各事象(1,2,3...)を変数と捉えてみる  
確率的に表れる変数であるため、これを**確率変数**と呼ぶ  
  
サイコロを振ってXが現れる確率を表にまとめると次のようになる  
  
  
この、各事象と、その出現確率を表したものが**確率分布**と呼ばれる  
Xを横軸に、Yを縦軸に取ればこれはグラフとしても描ける  
  
具体的には次のような式で表すことが出来る  
  
$P(X=3) = \frac{1}{6}$  
  
Xは現れるであろう事象のどれかを示す変数であるため、特定の事象を指す場合は、  
$X=3$ や $X_3$ のように表現する

### 期待値
仮にサイコロを振って、$出た目の数 \times 1000円$というゲームがあったとする  
これで平均的にどれくらいのお金がもらえるだろうか  
  
この、**確率分布に対する平均の事を期待値という**  
通常の算術平均は、全体の個数で総和を割る  
これは、得られたデータが母集団とみなされる一様分布における期待値と考えても良い  
  
期待値は英語でExpected Valueであるため、Xの期待値は$E(X)$のようにあらわされる  
  
期待値は次のように定義される  
  
$E(X) = \Sigma_{i=1}^n x_ip_i$  
  
これは、**各事象と、その確率の積の総和**と言える  
算術平均の場合$p_i$に対応するのが$\frac{1}{n}$であるため、一様分布と考えてよい  
  
サイコロを振って、$出た目の数 \times 1000円$というゲームの場合、  
$各目\times 1000円$が出る確率はそれぞれ$\frac{1}{6}$であるため  
  
$E(X) = (1000 + 2000 + \dots + 6000)\times \frac{1}{6} = 3500$  
  
つまり、このゲームをやると平均3500円ほど貰えるという事になる  

### 分散
上記では、確率分布における平均。つまり期待値を求めた  
同様に、分散も求める事が出来る  
分散が分かれば標準偏差も分かる  
  
分散は英語でVarianceであるため、確率変数Xの分散は$V(X)$で表される  
標準偏差は慣習的に$\sigma$が用いられる  
  
$V(X)=E((X-E(X))^2)$  
 
分散とは、平均偏差の平方和の平均だった  
上記の式を見ると  
  
$各値 - 期待値$  
これはつまり、平均偏差  
  
$(各値 - 期待値)^2$  
これは、平均偏差の平方  
  
$E((各値-期待値)^2)$  
Eという関数には、平均を求める処理が詰まっているため、これで、平均偏差の平方和の平均を求める事が出来る  
  
つまり、これで分散となる  
形は違うが内容は全く同じ  
  
同様に標準偏差は  
  
$\sigma (x) = \sqrt{V(X)}$  
  
で求められる
