確率変数と、確率分布の考え方を用いれば、変数と確率の対応を関数で捉える事が可能  
  
関数であるという事は、引数をとるという事  
引数が限定的な場合(例えばサイコロ)と、引数が自由な場合(連続的という事)で関数の呼ばれ方が異なる  
  
離散的な値を引数に取る確率関数 -> **確率質量関数**  
連続的な値を引数に取る確率関数 -> **確率密度関数**  
  
以下では、それぞれの確率関数の例を挙げる

### 離散一様分布
サイコロのように、それぞれの事象が現れる確率が同様で、一様であるような確率分布  
サイコロでは1～6だが、これをa～bで一般化すると次のようになる  
  
$
f(x) = 
  \left\{
    \begin{array}{l}
      \frac{1}{n} & a\leq x\leq b \\
      0 & otherwise \\
    \end{array}
  \right.
$  
  
これから、期待値は  
  
$E(x) = \Sigma_a^b (x_i\frac{1}{n}) = \frac{1}{n}\Sigma_a^b x_i$  
 
この時、$\Sigma_a^b x_i$は、a～bまでの差が1ずつ並んだ数列の和(等差数列の和)であると考えられるため  
https://manabitimes.jp/math/1120  
上記の証明を利用する(今回は公差が1の限定バージョン)  
a～bまでの和を次の二通り考える  
  
$S = a + (a+1) + \dots + b$  
$S = b + (b-1) + \dots + a$  
  
ここでこれら二つを足して、2で割ったものを考える  
  
$\frac{2S}{2} = S$  
$2S = (a+b) + (a+1 + b-1) + \dots + (b+a) = n(a+b)$  
$\frac{2S}{2} = \frac{n(a+b)}{2}$  
  
これから期待値は  
$E(x) = \frac{1}{n} \times \frac{n(a+b)}{2} = \frac{a+b}{2}$  
となる  
  
サイコロの場合a=1,b=6であるため  
$E(x) = \frac{1+6}{2} = 3.5$  
となる

### 正規分布
連続的な値を引数に取る、確率分布の事を**確率密度関数**と呼んだ  
その代表的なものが正規分布である  
以下で定義される  
  
$f(x) = \frac{1}{\sqrt{2\pi}}e^{-\frac{x^2}{2}}$  
  
以下のようなグラフが描ける  
  
  
  
xは$-\infty\leq x \leq\infty$の範囲で値をとる  
x=0の時最も大きい値をとり、0から離れるにつれて値が0に漸近していく  
  
考案者のガウスにちなんでガウス分布とも呼ばれる  
  
ガウスは天体観測で得られる値において次のような式を見出した  
  
$観測された値 = 真の値 + 正規分布で表現される誤差$  
  
正規分布の嬉しい所は、誤差を微分可能な一つの式として置き換える事が出来るところ  
これが機械学習でのパラメータ調整にも効いてくる  
  
ただし、確率密度関数で注意したいのは**確率密度関数は単一の事象に対して確率を算出できない(するべきではない)**というところ  
  
上記の正規分布の式にx=0を代入しても、値は得られるが、意味としては**x=0になる確率ではない**  
確率密度関数は**xがある範囲内で出現する確率**を求める  
  
要するに、確率を面積で返してくるのが確率密度関数である  
xは無限に近い数を取り得るため、ぴったり0であることはないと考える  
  
確率密度関数で、確率を得る時は積分を用いる  
  
$P(a\leq X \leq b) = \int_a^b f(d)dx$  
  
正規分布の場合  
  
$P(1\leq X \leq \infty) = \int_1^{\infty} \frac{1}{\sqrt{2\pi}}e^{-\frac{x^2}{2}} dx = 0.1586...$  
  
という具合に算出される  
これはつまり  
  
こういう範囲の面積を求めている  
正規分布において1以上の値が検出される確率は大体16%程度ということ  
  
また、  
$f(x) = \frac{1}{\sqrt{2\pi}}e^{-\frac{x^2}{2}}$  
上記の式で表される正規分布は  
期待値 0  
分散 1  
の特別な正規分布 **標準正規分布** である  
  
これを  
期待値 $\mu$  
分散 $\sigma^2$  
で一般化すると  
  
$f(x) = \frac{1}{\sqrt{2\pi\sigma^2}}e^{-\frac{(x-\mu)^2}{2\sigma^2}}$  
  
となる
  
$\mu$は正規分布の頂点の位置を、  
$\sigma$は正規分布の広がり方を決める
