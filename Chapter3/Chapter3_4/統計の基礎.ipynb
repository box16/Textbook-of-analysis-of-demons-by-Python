### 代表値
[総務統計局](http://www.stat.go.jp/data/kakei/5.html)から入手した情報を基に、統計の基礎を学ぶ  
  
  
#### 最小値(minimum)
得られたデータの中で最も小さな値の事  
  
#### 最大値(maximum)
得られたデータの中で最も大きな値の事  
  
#### 平均値(mean)
**算術平均**と呼ばれるもの  
データの個数の逆数を平等に重みとして加算した平均  
$\frac{1}{n}\Sigma_{i=1}^n x_i$  
  
#### 中央値(median)
データをソート(小さい順、大きい順)に並べたときに、ちょうど真ん中に来る値  
データが奇数の時と、偶数の時で、算出方法が異なる  
$中央値 = 
\left 
    \{ \begin{array}{ll}
        x_{\frac{n+1}{2}} & nが奇数 \\
        \frac{1}{2}(x_{\frac{n}{2}}+x_{\frac{n+1}{2}}) & nが偶数 \\ 
    \end{array}
\right.$  
  
#### 最頻値(mode)
得られたデータの中で、最も頻繁に出現するデータ  
  
#### 分位数
パーセンタイルや、クォンタイルとも呼ばれる  
中でも**四分位数**と呼ばれるものが最も使われる  
中央値と似た考え方で、  
全体の1/4番目に来るものを第一四分位数,  
全体の2/4番目に来るものを第二四分位数,  
全体の3/4番目に来るものを第三四分位数と呼ぶ  
箱ひげ図を作成する際に用いられる  
  
### バラツキの指標
平均値や中央値は、位置を示す代表値である  
しかし、これだけではロスする情報が多い  
位置を示す情報の他に、**バラツキ**も重要な指標となる  
  
例えば、四分位数の第三四分位数と第一四分位数はバラツキの指標として用いられる  
この範囲の事を**四分位範囲(IQR : Interquaritile Range)**と呼ぶ  
  
他の、バラツキの指標としては**分散と標準偏差**が良く用いられる  
分散とは、平均値$\bar x$を求めた後に次のようにして求める事が出来る  
  
$分散 = \frac{1}{n} \Sigma_{i=1}^n (x_i - \bar x)^2$  
  
つまり、平均値からの偏差の二乗(符号をそろえるための二乗)の平均値が分散である  
そして、二乗という操作によって単位が変わるため、これを打ち消すために標準偏差を用いる  
$標準偏差 = \sqrt{分散}$  
  
分散は、限られた情報(標本)から、母集団を推定する際には**不偏分散**として用いられることが多い  
この場合、  
$不偏分散 = \frac{1}{n-1} \Sigma_{i=1}^n (x_i - \bar x)^2$  
となる  
n-1の事を自由度と呼ぶ  
numpy,pandasではddof(delta degrees of freedom)として1を指定することで不偏分散の算出が出来る  

### 度数分布表
得られたデータの見通しを良くするために度数分布表というのが良く用いられる  
  
  
度数分布表の作成に良く用いられる方法としては、  
1. データの最大値と最小値を求める  
2. それらの間を等間隔に区切る(区切られた1区間の事を**階級**と呼ぶ)
3. 得られたデータそれぞれが、どの階級に属するのかカウントする(階級ごとに属する数を**度数**と呼ぶ) 
  
階級の作成には、これといった決まりがなく、データ可視化の際に微調整が必要となる  
少なすぎても、多すぎても、分布の形が分かりにくくなってしまう
