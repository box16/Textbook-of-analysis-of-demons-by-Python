  
例えば、上記の図で言うと、  
納豆の消費が多い都道府県では、わかめの消費も多い  
という事が分かる  
  
では、これの関係性の強さはどれくらいなのかと数値化してみたくなる  
それを実現するのが**相関係数**である  

### 相関係数
二軸で表されるデータ群に対して、**共分散**というものを考える事が出来る  
これは、一つの変数で表されるデータ群で算出される分散に近い考え方のものである  
以下の定義で共分散は算出される  
  
$s_{xy} = \frac{1}{n} \Sigma_{i=1}^n (x_i-\bar x)(y_i - \bar y)$  
  
ちなみに分散は以下の式で表される  
  
$s_x^2 = \frac{1}{n} \Sigma_{i=1}^n (x_i-\bar x)^2$  
  
共分散はまず、x軸、y軸それぞれの平均を求める  
各データがx,yの平均に対してどういう位置にいるのかを算出する  
* x,yともに平均より大きい → 正の値
* x,yともに平均より小さい → 正の値  
* x,yどちらかが平均より小さい → 負の値  
  
つまり、得られた散布図を4つの領域(2つの意味。正か負か)に分類し、それの平均を求める  
共分散に関しては**負の値も0もとり得る**  
  
そして、共分散と分散(標準偏差)を用いて相関係数$r_{xy}$は次のように定義される  
  
$r_{xy} = \frac{s_{xy}}{s_xs_y}$  
  
共分散を、xとyそれぞれの標準偏差の積で割ったものが相関係数となる  
  
相関係数は-1～1までの値をとる  
  
$-1 \leq r_{xy} \leq 1$  
  
-1に近いほど強い負の相関(右肩下がり)  
1に近いほど強い正の相関(右肩上がり)となる

### その他の相関係数
  
$r_{xy} = \frac{s_{xy}}{s_xs_y}$  
  
上記で定義される相関係数は**ピアソンの積率相関係数**と呼ばれるもの  
他に有名なものとしては**スピアマンの順位相関係数**というものがある  
  
$\rho_{xy} = 1 - \frac{6\Sigma d_i^2}{n(n^2-1)}$  
  
これは、データの順番に注目して算出される相関係数である  
つまり、離散的な値同士の相関係数(アンケート回答などに有効)である  
  
$d_i$はサンプルiにおける二つのデータの差を示す  
例えば、香川県は納豆の購入金額が全国で44位、わかめが26位。つまり$d_i = 44-26 = 18$  
各都道府県でこれの二乗を算出し、足し合わせるという事になる  
基本的な考え方は分散と同じ
