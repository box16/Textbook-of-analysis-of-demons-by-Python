# ベクトルとその演算

## ベクトル
将棋盤のように、縦、横で位置を指定するような仕組み  
$(4,7)$  
この場合、二つの軸の数字をひとまとまりに表現している  
  
ベクトルは矢印で表現できる  
  
原点を始点とするベクトルを**位置ベクトル**と呼ぶ  
$(4,7)$は位置を示すベクトルと捉える事が出来る  
  
ベクトルは向きと長さが同じであればよいため、原点始まりでなくともよい  
空間での位置を示すベクトルは横書きで、  
向きと大きさを示す位置ベクトルを縦書きで、描き分ける方法がある  
$\vec{x}= 
\left ( 
    \begin{array}{c}
        x_1\\
        x_2\\
        \vdots    \\
        x_n
    \end{array}
\right )
$

## ベクトルの演算
  
### ベクトルの足し算
ベクトルの足し算は対応する要素同士の足し算となる  
そのため、**同じ要素数のベクトル同士でしか足し算出来ない**  
$\vec{x} + \vec{y} = 
\left ( 
    \begin{array}{c}
        x_1 + y_1 \\
        x_2 + y_2 \\
        \vdots    \\
        x_n + y_n 
    \end{array}
\right )
$  
ベクトル同士の足し算は$\vec{x}と\vec{y}$が作る平行四辺形の対角線を作成するのと同じ意味になる  
つまり、ある始点からxの示す方向に、示す量だけ進み、yの示す方向に、示す量だけ進んだ位置を示している  


### ベクトルとスカラーの掛け算
ベクトルは複数の数をひとまとまりに表したもの  
対して、一つしかない数字の事を**スカラー**と呼ぶ  
  
ベクトルとスカラーは掛け算が出来る  
ベクトルのそれぞれの要素をスカラー倍する  
$a\vec{x}= 
\left ( 
    \begin{array}{c}
        ax_11 \\
        ax_2  \\
        \vdots    \\
        ax_n 
    \end{array}
\right )
$  
-1を掛けた場合は、ベクトルの向きが反転するのと同じ

### ベクトルの引き算
基本的な考え方は足し算と同じ  
意味するところとしては、対角線から、平行四辺形の一辺を求めるのと同じ  
$\vec{x} - \vec{y} = 
\left ( 
    \begin{array}{c}
        x_1 - y_1 \\
        x_2 - y_2 \\
        \vdots    \\
        x_n - y_n 
    \end{array}
\right )
$  
  

### ノルム
ベクトルの大きさをスカラーで表現する時これを**ノルム**と呼ぶ  
ノルム(norm)は英語で**標準、基準**を意味する  
  
最も一般的なベクトル->スカラー変換は以下  
$\|x\| = \sqrt{x_1^2 + x_2^2 + \dots + x_n^2 }$  
これによって算出されるスカラーを**ユークリッド距離**とも呼ぶ  
始点と、終点を一直線で結ぶ直線距離がこれで分かる  
  
もう一つ、ベクトルの各要素の絶対値を足し合わせるという方法も考えられる  
$\|x\|_1 = |x_1| + |x_2| \dots + |x_n|$  
これによって、算出されるスカラーを**マンハッタン距離**と呼ぶ  
ユークリッド距離は始点と終点を結ぶ一直線の長さだったが、  
マンハッタン距離は、軸に沿いながら目的地を目指すときの距離  
  
ユークリッド距離とマンハッタン距離を区別するために以下のように描き分けることもある  
$ユークリッド距離 : \|x\|_2 (別名:L^2ノルム) \\
マンハッタン距離 : \|x\|_1 (別名:L^1ノルム)$　　
  
この関係を一般化して、$L^pノルム$を考える事が出来る  
$\|x\|_p = (|x_1|^p + |x_2|^p + \dots + |x_n|^p)^{\frac{1}{p}}$

### 内積
ベクトル同士の掛け算には2種類ある  
そのうちの一つが**内積**  
内積の結果はスカラーになる  
  
$ \vec{x} \cdot \vec{y} = \Sigma_i^n x_iy_i = x_1y_1 + x_2y_2 + \dots + x_ny_n $
  
また、内積を二つのベクトルの$L^2ノルム$で割ると2つのベクトルがなす角度の$\cos$になる  
$\cos\theta = \frac{\vec{x}\cdot\vec{y}}{\|x\|\|y\|}$  
  
ベクトルの掛け算は足し算程簡単にはいかない  
足し算は矢印の先に別の矢印を付ける、もしくは取り除くことで実現される  
掛け算の場合、直感的に操作ができないため、まず**ベクトル同士を同じ軸に落とし込む**
